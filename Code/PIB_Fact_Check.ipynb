```
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```
def clean_text(text):
  text = text.lower()
  text = re.sub(r'http\S+|www\S+', '', text)
  text = word_tokenize(text)
  text = [t for t in text if t not in stop_words]
  text = [t for t in text if t.isalpha()]
  text = [lemmatizer.lemmatize(t) for t in text]
  text = ' '.join(text)
  return text
```

```
clean_text(df.text[0])
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
corpus = df.text
tfidf_vectorizer = TfidfVectorizer(preprocessor=clean_text)
tfidf_data = tfidf_vectorizer.fit_transform(corpus)
```

```
tfidf_vectorizer.get_feature_names()[:10]
```

```
terms = tfidf_vectorizer.get_feature_names()

# sum tfidf frequency of each term through documents
sums = tfidf_data.sum(axis=0)

# connecting term to its sums frequency
data = []
for col, term in enumerate(terms):
    data.append(( term, sums[0,col] ))

ranking = pd.DataFrame(data, columns=['term','rank'])
ranking = ranking.sort_values('rank', ascending=False)
```

```
print(ranking[:50])
```

```
def sort_coo(coo_matrix):
    tuples = zip(coo_matrix.col, coo_matrix.data)
    return sorted(tuples, key=lambda x: (x[1], x[0]), reverse=True)

def extract_topn_from_vector(feature_names, sorted_items, topn=10):
    """get the feature names and tf-idf score of top n items"""
    
    #use only topn items from vector
    sorted_items = sorted_items[:topn]

    score_vals = []
    feature_vals = []

    for idx, score in sorted_items:
        fname = feature_names[idx]
        
        #keep track of feature name and its corresponding score
        score_vals.append(round(score, 3))
        feature_vals.append(feature_names[idx])

    #create a tuples of feature,score
    #results = zip(feature_vals,score_vals)
    results= {}
    for idx in range(len(feature_vals)):
        results[feature_vals[idx]]=score_vals[idx]
    
    return results
```

```
feature_names = tfidf_vectorizer.get_feature_names()
sorted_items = sort_coo(tfidf_data.tocoo())
docs = df.text.tolist()
results=[]
for i in range(tfidf_data.shape[0]):
    
    # get vector for a single document
    curr_vector=tfidf_data[i]
    
    #sort the tf-idf vector by descending order of scores
    sorted_items=sort_coo(curr_vector.tocoo())

    #extract only the top n; n here is 10
    keywords=extract_topn_from_vector(feature_names,sorted_items,10)
    
    
    results.append(keywords)

docs_tfidf=pd.DataFrame(zip(docs,results),columns=['doc','keywords'])
docs_tfidf
```
